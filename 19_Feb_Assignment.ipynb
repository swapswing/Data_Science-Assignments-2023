# Q1. What is Flask Framework? What are the advantages of Flask Framework?

Ans.Flask is a lightweight web framework written in Python. It is designed with simplicity and flexibility in mind, making it easy to build web applications quickly and with less boilerplate code compared to other frameworks like Django.

Advantages of Flask Framework:

Simplicity: Flask has a simple and easy-to-understand API.
Flexibility: It allows developers to choose the tools and libraries they want to use.
Extensibility: Flask can be extended with various Flask extensions for specific functionalities.
Jinja2 Templating: Built-in support for powerful Jinja2 templating engine.
Werkzeug: Integrated with Werkzeug, a comprehensive WSGI utility library.

# Q3. What is App routing in Flask? Why do we use app routes?

Ans.App routing in Flask refers to the mechanism by which URLs are mapped to functions. In other words, when a user visits a particular URL on your Flask application, Flask determines which Python function should handle that request. This is achieved using the @app.route() decorator in Flask.

Why do we use app routes?

Organize URLs: Routes help organize different parts of your application.
Separation of Concerns: Routes separate different functionalities or views into distinct parts.
Readability: URLs become more readable and descriptive.
Ease of Maintenance: Makes it easier to modify or add new functionality to specific URLs.
